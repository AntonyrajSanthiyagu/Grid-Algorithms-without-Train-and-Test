```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
indep=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dep=dataset['charges']
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression
param_grid = {'fit_intercept': [True, False],'copy_X': [True, False],'n_jobs': [None, 1, -1],'positive': [False, True]}
grid=GridSearchCV(LinearRegression(), param_grid, refit=True,cv=5,verbose=3, n_jobs=-1)
grid.fit(indep,dep)
```

```python
Y_pred=grid.predict(indep)
```

```python
#print best parameter after tuning
#print(grid.best_params_)
re=grid.cv_results_
from sklearn.metrics import r2_score
r_score=r2_score(dep,Y_pred)
print("The R_score value for best parameter {}:".format(grid.best_params_),r_score)
```
